#### 3.7 Adding columns for "Test Score" and "Average Score"

```python
df.head()
```

```python
df["Total_score"] = df["math score"]+df["reading score"]+df["writing score"]
df["avg_score"] = df["Total_score"]/3
df.head() # "Test Score" and "Average Score" are dependent feature and rest all the columns are independent features
```

```python
reading_full = df[df["reading score"] == 100]["avg_score"].count()
writing_full = df[df["writing score"] == 100]["avg_score"].count()
math_full = df[df["math score"] == 100]["avg_score"].count()

print(f"Number of students with full marks in Maths: {math_full}")
print(f"Number of students with full marks in Maths in writing: {writing_full}")
print(f"Number of students with full marks in Maths in reading: {reading_full}")
```

```python
reading_less_20 = df[df["reading score"] <= 20]["avg_score"].count()
writing_less_20 = df[df["writing score"] <= 20]["avg_score"].count()
math_less_20 = df[df["math score"] <= 20]["avg_score"].count()

print(f"Number of students with full marks in Maths: {math_less_20}")
print(f"Number of students with full marks in Maths in writing: {writing_less_20}")
print(f"Number of students with full marks in Maths in reading: {reading_less_20}")
```

##### Insights

* From above values we get students have performed the worst in Maths.
* Best performance is in reading section.

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64'])
gp = numeric_columns.groupby(df['gender'])
gender_group = gp.mean()
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average']


female_scores = [gender_group['avg_score'][0], gender_group['math score'][0]]
male_scores = [gender_group['avg_score'][1], gender_group['math score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

##### Insights
* On an average females have a better overall score than men.
* whereas males have scored higher in Maths.
